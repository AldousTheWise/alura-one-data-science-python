2.- Con los mismos datos de la pregunta anterior, determina cuántas compras se realizaron por encima de 3000 reales y calcula el porcentaje con respecto al total de compras.

````python
# Inicializamos una variable como contador.
compras_grandes = 0;

# Hacemos un ciclo for que recorra el array.
for gasto in gastos:
    if gasto > 3000: # Si en el array se encuentra un valor superior 3000
        compras_grandes += 1; # Actualiza el valor del contador.

total_compras = len(gastos); # Establecemos una variable que devuelva el tamaño del array.

porcentaje = (compras_grandes / total_compras) * 100;

print(f"Cantidad de compras > 3000: {compras_grandes}");
print(f"Porcentaje de compras > 3000: {porcentaje:.2f}%");
````

5.- Crea un programa que, al ingresar un número cualquiera, genere una lista que contenga todos los números primos entre 1 y el número ingresado.

````python
# Se pide un número al usuario:
num = int(input("Ingresa un número: "))

# Lista para guardar los nros. primos:
primos = [];

# Recorremos del 2 al nro. ingresado (recuerda que el 1 no se considera número primo).
for n in range(2, num + 1):
    es_primo = True # Asumimos que el número ingresado es primo.
    print(f"\nProbando si {n} es primo");

    for divisor in range(2, n):
        print(f"\nProbando divisor {divisor}... ", end = "");
        if n % divisor == 0:
            print(f"{n} es divisible por {divisor}.");
            es_primo = False; # Si tiene divisor, no es primo.
            break;
        else:
            print("no lo divide");

    if es_primo:
        primos.append(n); # Agregamos los números al array.
        print(f"{n} es primo");
    else:
        print(f"{n} no es primo");

# Mostrar resultado:
print(f"\nNúmeros primos 1 y {num}: {primos}");
````

6.- Escribe un programa que pida una fecha, especificando el día, mes y año, y determine si es válida para su análisis.

````python
# Pedimos los datos:
day = int(input("Ingresa el día: "));
month = int(input("Ingrese el mes: "));
year = int(input("Ingresa el año: "));

# Lista con los días de cada mes; indice[0] es enero, indice[1] es febrero y asi sucesivamente...
dias_por_mes = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31];

# Ajuste de febrero si el año es bisiesto:
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    dias_por_mes[1] = 29;

# Validamos la fecha:
if month < 1 or month > 12:
    print("Mes inválido");
elif day < 1 or day > dias_por_mes[month - 1]: # Hacemos el ajuste [month - 1] para que cuadre con el número de índice en el array.
    print("Día inválido para el mes");
else:
    print(f"La fecha {day}/{month}/{year} es válida.");
````

7.- Para un estudio sobre la multiplicación de bacterias en una colonia, se recopiló el número de bacterias multiplicadas por día y se puede observar a continuación:

```bash 
[1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]
``` 
Con estos valores, crea un código que genere una lista que contenga el porcentaje de crecimiento de bacterias por día, comparando el número de bacterias en cada día con el número de bacterias del día anterior.  
**Sugerencia:** para calcular el porcentaje de crecimiento, utiliza la siguiente ecuación: 

```bash
100 * (muestra_actual - muestra_anterior) / muestra_anterior
```

````python
# Datos de la colonia:
bacterias = [1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9];

# Array para los porcentajes de crecimiento:
crecimiento = [];

# Recorremos desde el segundo día (indice 1):
for i in range(1, len(bacterias)):
    actual = bacterias[i];
    anterior = bacterias[i - 1];
    porcentaje = 100 * (actual - anterior) / anterior;
    crecimiento.append(porcentaje);

# Mostramos resultados:
for i, pct in enumerate(crecimiento, start=2): # start=2 porque el primer dia de crecimiento es el dia 2.
    print(f"Dia {i}: Crecimiento {pct:.2f}");
````

8.- Para una selección de productos alimenticios, debemos separar el conjunto de IDs proporcionados por números enteros, sabiendo que los productos con ID par son dulces y los que tienen ID impar son amargos. Crea un código que recoja 10 IDs. Luego, calcula y muestra la cantidad de productos dulces y amargos.

````python
# Lista para guardar los ids:
id_productos = [];

# Pedimos 10 IDs al usuario:
for i in range(10):
    id_producto = int(input(f"Ingrese el ID del producto {i+1}:"));
    id_productos.append(id_producto);

# Contadores de dulces y amargos:
dulces = 0;
amargos = 0;

# Clasificamos los ids:
for id_producto in id_productos:
    if id_producto % 2 == 0: # par -> dulce
        dulces += 1;
    else: # impar -> amargo
        amargos += 1;

print(f"\nCantidad de productos dulces: {dulces}");
print(f"Cantidad de productos amargos: {amargos}");
````

9.- Desarrolla un programa que informe la puntuación de un estudiante de acuerdo con sus respuestas. Debe pedir la respuesta del estudiante para cada pregunta y verificar si la respuesta coincide con el resultado. Cada pregunta vale un punto y hay opciones A, B, C o D.

```bash
Resultado del examen:
01 - D
02 - A
03 - C
04 - B
05 - A
06 - D
07 - C
08 - C
09 - A
10 - B
```

````python
# Respuestas correctas:
respuestas_correctas = ['D', 'A', 'C', 'B', 'A', 'D', 'C', 'C', 'A', 'B'];

# Lista para guardar las respuestas del estudiante:
respuestas_estudiante = [];

# Pedimos las respuestas al estudiante:
for i in range(10):
    resp = input(f"Pregunta {i + 1}: ").upper(); # .upper() asegura que la respuesta sea en mayúscula.
    
    while resp not in ['A', 'B', 'C', 'D']: # Validación
        resp = input("Respuesta inválida. Ingrese A, B, C ó D: ").upper();
    
    respuestas_estudiante.append(resp); # Almacena las respuestas en respuestas_estudiante.

# Calculamos la puntuación:
puntaje = 0; 

for correcta, estudiante in zip(respuestas_correctas, respuestas_estudiante): 
    '''
    El método zip() toma dos o más arrays (o cualquier iterable) y las empareja
    elemento por elemento en tuplas.
    En este caso emparejaría: 
    respuestas_correctas[0] : respuestas_estudiante[0]
    respuestas_correctas[1] : respuestas_estudiante[1]
    ...
    respuestas_correctas[9] : respuestas_estudiante[9]
    sucesivamente.
    '''
    if correcta == estudiante:
        puntaje += 1;


# Mostramos resultado:
print(f"\nEl estudiante obtuvo {puntaje} puntos sobre 10");
````

10.- Un instituto de meteorología desea realizar un estudio de la temperatura media de cada mes del año. Para ello, debes crear un código que recoja y almacene esas temperaturas medias en una lista. Luego, calcula el promedio anual de las temperaturas y muestra todas las temperaturas por encima del promedio anual y en qué mes ocurrieron, mostrando los meses por su nombre (Enero, Febrero, etc.).

````python
# Lista con los meses:
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"];

# Lista para almacenar t°
temperaturas = [];

for mes in meses:
    temp = float(input(f"Ingrese la tempertatura promedio de {mes}: "));
    temperaturas.append(temp);

# Calculamos el promedio anual:
promedio_anual = sum(temperaturas) / len(temperaturas);

print(f"\nTemperatura promedio anual: {promedio_anual:.2f}°C\n");

# Mostramos las t° por encima del promedio:
print("Meses con temperatura por encima del promedio:");
for i in range(len(temperaturas)):
    if temperaturas[i] > promedio_anual:
        print(f"{meses[i]} -> {temperaturas[i]:.2f}°C");
````

12.- Se realizó una encuesta de mercado para decidir cuál diseño de marca infantil es más atractivo para los niños. Los votos de la encuesta se pueden ver a continuación:

```bash
Tabla de votos de la marca: 
Diseño 1 - 1334 votos
Diseño 2 - 982 votos
Diseño 3 - 1751 votos
Diseño 4 - 210 votos
Diseño 5 - 1811 votos
```

Adapta los datos proporcionados a una estructura de diccionario. A partir de ello, informa el diseño ganador y el porcentaje de votos recibidos.

````python
# Diccionario de votos:
votos = {
    "Diseño 1" : 1334,
    "Diseño 2" : 982,
    "Diseño 3" : 1751,
    "Diseño 4" : 210,
    "Diseño 5" : 1811
}

# Calcular el total de votos:
total_votos = sum(votos.values());

# Encontrar el diseño con más votos:
ganador = max(votos, key=votos.get); 
'''
El método max() extrae el mayor valor de una colección, en este caso 
del diccionario 'votos', sin embargo debemos decirle que solo compare
el valor de las claves.
En tal punto usamos la version abreviada del método get().
'''

# Calcular el porcentaje del ganador:
porcentaje = (votos[ganador] / total_votos) * 100;

# Mostrar resultados:
print(f"El diseño ganador es {ganador} con {votos[ganador]} votos.");
print(f"Porcentaje de votos: {porcentaje:.2f}% del total ({total_votos}).");
````

13.- Los empleados de un departamento de tu empresa recibirán una bonificación del 10% de su salario debido a un excelente rendimiento del equipo. El departamento de finanzas ha solicitado tu ayuda para verificar las consecuencias financieras de esta bonificación en los recursos. Se te ha enviado una lista con los salarios que recibirán la bonificación: 

```bash
[1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903].
``` 

La bonificación de cada empleado no puede ser inferior a 200. En el código, convierte cada uno de los salarios en claves de un diccionario y la bonificación de cada salario en el valor correspondiente. Luego, informa el gasto total en bonificaciones, cuántos empleados recibieron la bonificación mínima y cuál fue el valor más alto de la bonificación proporcionada.

````python
# Lista de salarios:
salarios = [1172, 1644, 2617, 5130, 5532, 6341, 6650, 7238, 7685, 7782, 7903]

# Creamos el diccionario: cada salario es una clave (key) y su bonificacion, el valor (value):
# Usamos max(..., 200) para asegurar que la bonificación mínima sea de 200.
bonificaciones = {salario: max(salario * 0.10, 200) for salario in salarios};

# Mostramos el diccionario de bonificaciones:
print("Tabla de bonificaciones:");
for salario, bono in bonificaciones.items():
    print(f"Salario: {salario} -> Bonificación: {bono:.2f}");

# Calculamos el gasto total (la suma de todos los valores):
gasto_total = sum(bonificaciones.values());

# Contamos cuantos empleados recibieron exactamente 200:
minimos = list(bonificaciones.values()).count(200);

# Identificamos la bonificación más alta:
bono_maximo = max(bonificaciones.values());

# Mostramos los resultados:
print("\nResumen:");
print(f"Gasto total en bonificaciones: ${gasto_total:.2f}");
print(f"Empleados con bonificación mínima: {minimos}");
print(f"Bonificación más alta: ${bono_maximo:.2f}");
````

15.- El departamento de Recursos Humanos de tu empresa te pidió ayuda para analizar las edades de los colaboradores de 4 sectores de la empresa. Para ello, te proporcionaron los siguientes datos:

```bash
{
    'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
    'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
    'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
    'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]
}
```

 Dado que cada sector tiene 10 colaboradores, construye un código que calcule la media de edad de cada sector, la edad media general entre todos los sectores y cuántas personas están por encima de la edad media general.

````python
# Información de edades por sector:
edades = {
    'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
    'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
    'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
    'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]
};

# Media por sector:
medias_sector = {};
for sector, lista in edades.items():
    media = sum(lista) / len(lista);
    medias_sector[sector] = media;

# Edad media general:
todas_edades = [];
for lista in edades.values():
    todas_edades.extend(lista);

media_general = sum(todas_edades) / len(todas_edades);

# Conteo de cuantos están por encima de la media general:
contador = 0;
for edad in todas_edades:
    if edad > media_general:
        contador += 1;

# Mostrar resultados:
print("Media de edad por sector:");
for sector, media in medias_sector.items():
    print(f"{sector} : {media:.2f} años");

print(f"\nEdad media general: {media_general:.2f} años");
print(f"Número de empleados por encima de la media general: {contador}");
````
