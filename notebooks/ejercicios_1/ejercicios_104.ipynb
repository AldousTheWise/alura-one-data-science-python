1 - Escribe un programa que solicite dos números enteros e imprima todos los números enteros entre ellos.

```python
num_inicial = int(input("Ingresa el primer número: "));
num_final = int (input("Ingresa el segundo número: "));


if num_final > num_inicial:
    for i in range(num_inicial + 1, num_final):
        print(i);
else:
    for i in range(num_inicial - 1, num_final, -1):
        print(i);
```

2 - Escribe un programa para calcular cuántos días tomará que la colonia de una bacteria A supere o iguale a la colonia de una bacteria B, basado en tasas de crecimiento del 3% y 1.5%, respectivamente. Supón que la colonia A comienza con 4 elementos y B con 10.

```python
# Muestreo de bacterias:
colonia_A = 4;
colonia_B = 10;

# Tasa de crecimiento:
tasa_A = 1.03;
tasa_B = 1.015;

# Contador de días:
dias = 0;

# Mientras A no sea mayor o igual a B:
while colonia_A < colonia_B:
    colonia_A *= tasa_A; # A crece 3%
    colonia_B *= tasa_B; # B crece 1.5%
    dias += 1;

print(f"La colonia A superará o igualará a la B en { dias } días.");
print(f"Tamaño final de A: {colonia_A:.2f}");
print(f"Tamaño final de B: {colonia_B:.2f}");
```

3 - Para procesar una cantidad de 15 datos de evaluaciones de usuarios de un servicio de la empresa, necesitamos verificar si las calificaciones son válidas. Por lo tanto, escribe un programa que recibirá calificaciones del 0 al 5 y verificará si son valores válidos. Si se ingresa una calificación superior a 5 o inferior a 0, se repetirá hasta que el usuario ingrese un valor válido.

```python
# Array para guardar las calificaciones:
calificaciones = [];

# Queremos 15 calificaciones válidas:
for i in range(15):
    while True: # Loop while verificando si condicion es true, que el usuario ingrese un valor válido.
        nota = float(input(f"Ingrese la calificación #{i+1} (0 a 5): "));
        if 0 <= nota <= 5: # Notación abreviada de AND, equivalente a (if nota <= 0 AND nota <= 5)
            calificaciones.append(nota);
            break; # Salimos del while si es válido.
        else:
            print("Ingrese calificación válida entre 0 y 5. Intente nuevamente.")

print("Las calificaciones ingresadas: ");
print(calificaciones);
```

4 - Desarrolla un programa que lea un conjunto indefinido de temperaturas en grados Celsius y calcule su promedio. La lectura debe detenerse al ingresar el valor -273°C.

```python
# Variables para sumar t° y contador de ingresos.
suma_temperaturas = 0;
contador = 0;

while True:
    temp = float(input("Ingrese T° en °C (-273 para terminar)"));

    if temp == -273: # Condición de parada.
        break;

    suma_temperaturas += temp; # Sumatoria de temperaturas.
    contador += 1;  # se aumenta el contador.

# Se calcula el promedio, evitando la division por cero.

if contador > 0:
    promedio = suma_temperaturas / contador;
    print(f"\nSe ingresaron {contador} temperaturas.");
    print(f"El promedio es: {promedio:.2f} °C");

else:
    print("\nNo se ingresaron temperaturas.");
```

5 - Escribe un programa que calcule el factorial de un número entero proporcionado por el usuario. Recuerda que el factorial de un número entero es el producto de ese número por todos sus antecesores hasta llegar al número 1. Por ejemplo, el factorial de 5 es 5 x 4 x 3 x 2 x 1 = 120.

```python
def factorial(n):
    if n < 0:
        return "El factorial no está definido para números negativos";
    elif n == 0 or n == 1:
        print(f"factorial({n}) = 1  (caso base)");
        return 1;
    else:
        # return n * factorial(n - 1) # La función se llama a sí misma.
        print(f"factorial({n}) = {n} * factorial({n-1})")
        resultado = n * factorial(n - 1)
        print(f"Resultado parcial: factorial({n}) = {resultado}")
        return resultado
    
# Programa principal
numero = int(input("Ingrese un número entero: "));
resultado = factorial(numero);
print(f"El factorial de {numero} es {resultado}.");
```

7 - Los números primos tienen diversas aplicaciones en Ciencia de Datos, como en criptografía y seguridad. Un número primo es aquel que es divisible solo por sí mismo y por 1. Por lo tanto, crea un programa que solicite un número entero y determine si es un número primo o no.

```python
num = int(input("Ingrese un número entero: "));

if num <= 1: # Nos aseguramos que el número no sea 1 o menor que 1.
    print(f"{num} no es un número primo.")

else:
    es_primo = True; # Asumimos que el número es primo hasta que se demuestre lo contrario.

    for i in range(2, int(num ** 0.5) + 1): # Comprobamos hasta la raíz cuadrada.
        if num % i == 0:
            es_primo = False;
            break;

if es_primo:
    print(f"{num} es un número primo");
else:
    print(f"{num} no es un número primo.")
```

8 - Vamos a comprender la distribución de edades de los pensionistas de una empresa de seguros. Escribe un programa que lea las edades de una cantidad no informada de clientes y muestre la distribución en los intervalos [0-25], [26-50], [51-75] y [76-100]. La entrada de datos se detendrá al ingresar un número negativo.

```python
# Definimos los límites de rangos:
limites = [25, 50, 75, 100];
rangos = [0] * len(limites); # Crea una lista de contadores dentro de un array con cuatro ceros. [0, 0, 0, 0]

while True:
    edad = int(input("Ingresa la edad (Negativo para salir)"));
    
    if edad < 0:
        break;

    # Recorremos los límites para ver dónde encaja la edad:
    asignada = False;

    for i, limite in enumerate(limites):
        if edad <= limite:
            rangos[i] += 1;
            asignada = True;
            break;

    if not asignada:
        print("Edad fuera del rango esperado (0-100).");

# Mostrar resultados:
print("\nDistribución de edades:")
inicio = 0;
for i, limite in enumerate(limites):
    cantidad = rangos[i];
    barra = "#" * cantidad; # Tamaño de la barra
    print(f"[{inicio}-{limite}]: {cantidad} {barra}");
    inicio = limite + 1;
```

9 - En una elección para la gerencia de una empresa con 20 empleados, hay cuatro candidatos. Escribe un programa que calcule al ganador de la elección. La votación se realizó de la siguiente manera:

Cada empleado votó por uno de los cuatro candidatos (representados por los números 1, 2, 3 y 4).

También se contaron los votos nulos (representados por el número 5) y los votos en blanco (representados por el número 6).

Al final de la votación, el programa debe mostrar el número total de votos para cada candidato, los votos nulos y los votos en blanco. Además, debe calcular y mostrar el porcentaje de votos nulos con respecto al total de votos y el porcentaje de votos en blanco con respecto al total de votos.

```python
# Inicializamos contadores:
votos = [0, 0, 0, 0];
nulos = 0;
blancos = 0;
total_votantes = 20;

# Recolección de votos:
for i in range(1, total_votantes + 1):
    voto = int(input(f"Empleado {i}, ingrese su voto :\n(1-4) candidatos.\n(5) nulo.\n(6) blanco."));

    if 1 <= voto <= 4:
        votos[voto - 1] += 1;
    elif voto == 5:
        nulos += 1;
    elif voto == 6:
        blancos += 1;
    else:
        print("Voto inválido. No se cuenta.");


# Resultados:
print("\nResultados finales:")
for i, cant in enumerate(votos, start=1):
    print(f"Candidato {i}: {cant} votos");
print(f"Votos nulos: {nulos}");
print(f"Votos en blanco: {blancos}");

# Porcentajes:
porc_nulos = (nulos / total_votantes) * 100;
porc_blancos = (blancos / total_votantes) * 100;
print(f"Porcentaje de votos nulos: {porc_nulos:2f}%");
print(f"Porcentaje de votos en blanco: {porc_blancos:.2f}%");

# Determinar ganador:
max_votos = max(votos);
ganadores = [i + 1 for i, v in enumerate(votos) if v == max_votos];

if len(ganadores) == 1:
    print(f"\nEl ganador es el candidato {ganadores[0]} con {max_votos} votos");
else:
    print(f"\nEmpate entre los candidatos: {ganadores} con {max_votos} votos cada uno.");





```
