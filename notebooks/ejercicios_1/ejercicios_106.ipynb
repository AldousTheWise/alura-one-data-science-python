1.- Bruno gestiona un pequeño comercio y quiere saber qué producto tuvo el mejor desempeño de ventas el mes pasado. Registró la cantidad vendida de dos productos: manzanas y plátanos. Ahora, necesita escribir un programa que identifique y muestre cuál de ellos tuvo más ventas.

Crea un programa que reciba el número de ventas de los dos productos y muestre un mensaje indicando cuál de ellos vendió más. Si las cantidades son iguales, muestra un mensaje diciendo que hubo un empate.

````python
cant_manzanas = int(input("Digite la cantidad de manzanas vendidas: "));
cant_platanos = int(input("Digite la cantidad de plátanos vendidos: "));

if cant_manzanas > cant_platanos:
    print("Las manzanas tuvieron más ventas");
elif cant_manzanas < cant_platanos:
    print("Los plátanos tuvieron más ventas");
else:
    print("Ambas frutas tuvieron la misma cantidad de ventas."); 
````

2.- Lucas trabaja en TI y necesita garantizar que la temperatura de una sala de servidores no supere los 25°C. Quiere un programa que reciba la temperatura actual como entrada y, si es necesario, muestre un mensaje de alerta.

````python
try:
    temperatura = float(input("Digite la temperatura actual (formato x.x):"));
    if temperatura > 25.0:
        print(f"Alerta! Temperatura {temperatura}°C sobre límites recomendables.");
    else:
        print(f"Temperatura {temperatura}°C dentro de márgenes recomendados.")
except ValueError:
    print("Ingrese un número válido");
````

3.- Camila está organizando un proyecto y necesita calcular el tiempo total necesario para concluir tres actividades: A, B y C. Sin embargo, si alguna actividad tiene un número de días negativo, el código debe avisar que los valores ingresados son inválidos y no calcular el total.

Escribe un programa que reciba el número de días de tres actividades y muestre el tiempo total del proyecto. Si algún valor es negativo, muestra un mensaje informando el error.

````python
# Se define una función usando while para validar que el número ingresado sea un número válido (positivo).
def pedir_dias(actividad):
    while True:
        try:
            dias = int(input(f"Informe los días para la actividad {actividad}: "));
            if dias < 0:
                print("Error: Los días no pueden ser negativos.");
            else:
                print(f"La actividad {actividad} toma {dias} días.");
                return dias;
        except ValueError:
            print("Error: debes ingresar un número válido.");

# Pedimos los días de cada actividad de forma dinámica:
actividad_A = pedir_dias("A");
actividad_B = pedir_dias("B");
actividad_C = pedir_dias("C");

# Se calcula el total:
total_dias = actividad_A + actividad_B + actividad_C;
print(f"El tiempo total del proyecto es: {total_dias} días.");
````

4.- Anna Julia está creando un sistema para calcular el Índice de Masa Corporal (IMC) y proporcionar recomendaciones básicas. El programa debe recibir el peso y la altura de una persona y mostrar el valor del IMC, además de indicar si está por debajo del peso, con peso normal o por encima del peso.  
Crea un programa que reciba el peso (en kg) y la altura (en metros) y calcule el IMC usando la fórmula: 
```bash 
IMC = peso / (altura ** 2)
```
Luego, muestra el valor del IMC y un mensaje indicando si está por debajo del peso (IMC < 18.5), peso normal (18.5 <= IMC < 25) o por encima del peso (IMC >= 25).

````python
peso = float(input("Digite su peso: "));
altura = float(input("Digite su altura (ej: 1.75)"));

IMC = int(peso/(altura ** 2));

if IMC >= 25:
    print(f"Su IMC es: {IMC}.\nTienes sobrepeso.");
elif 18.25 <= IMC < 25:
    print(f"Su IMC es: {IMC}.\nTienes un peso normal.");
else:
    print(f"Su IMC es: {IMC}.\nTienes un peso bajo.");
````

7.- Estás desarrollando un pequeño juego. El usuario ingresa un número entero y el programa debe evaluar lo siguiente:

- Si el número es divisible por 3 y 5, muestra: "¡Número mágico!".
- Si solo es divisible por 3, muestra: "Divisible por 3".
- Si solo es divisible por 5, muestra: "Divisible por 5".
- Si no es divisible por ninguno, muestra: "No es un número mágico".

Este tipo de lógica es muy útil en juegos, validaciones o filtros.

````python
num = int(input("Ingrese un número:"));

if num % 5 == 0 and num % 3 == 0:
    print("¡Número mágico!");
elif num % 5 == 0:
    print("Divisible por 5");
elif num % 3 == 0:
    print("Divisible por 3");
else:
    print("No es un número mágico");
````

Una empresa evalúa su trimestre con base en:
- Ingresos totales
- Gastos totales
- Número de nuevos clientes

Clasificación:
- Si ingresos - gastos > $10,000 y más de 50 nuevos clientes → "Trimestre Excelente"
- Si ingresos - gastos > $5,000 y al menos 20 clientes → "Trimestre Bueno"
- Si ingresos - gastos > 0 → "Trimestre Regular"
- Si ingresos - gastos ≤ 0 → "Trimestre Deficitario"

````python
ingresos = int(input("Digite la cantidad de ingresos trimestrales:"));
gastos = int(input("Ingrese la cantidad de gastos totales en el trimestre:"));
clientes = int(input("Digite la cantidad de nuevos clientes en el trimestre:"));

diferencia = ingresos - gastos;

if diferencia > 10000 and clientes > 50:
    print("Trimestre excelente.");
elif diferencia > 5000 and clientes >= 20:
    print("Trimestre bueno.");
elif diferencia > 0:
    print("Trimestre regular.");
else:
    print("Trimestre deficitario");
````
