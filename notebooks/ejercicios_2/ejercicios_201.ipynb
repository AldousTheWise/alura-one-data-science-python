6 - Se debe escribir un programa para sortear a un seguidor de una red social para ganar un premio. La lista de participantes está numerada y debemos elegir aleatoriamente un número según la cantidad de participantes. Pide a la persona usuaria que proporcione el número de participantes del sorteo y devuelve el número sorteado.

````python
import random
````

````python
participantes = int(input("Ingrese la cantidad de participantes:"))

# Sorteo (entre 1 y la cantidad de participantes)
ganador = random.randint(1, participantes)

print(f"El número ganador es:  {ganador}")
````

````python
# Variante con nombres

cant = int(input("¿Cuántos participantes hay?"))
participantes = []

for i in range(cant):
    nombre = input(f"Ingrese el nombre o usuario del participante #{i+1}")
    print(f"Usuario {i+1} ingresado: {nombre}")

    participantes.append(nombre);

ganador = random.choice(participantes)

print(f"\nEl ganador del sorteo es: {ganador}")
````

9 - Has recibido un desafío para calcular la raíz cuadrada de una lista de números, identificando cuáles resultan en un número entero. La lista es la siguiente:

```bash
numeros = [2, 8, 15, 23, 91, 112, 256]
```

````python
import math
````

````python
numeros = [2, 8, 15, 23, 91, 112, 256]
enteros = []

for n in numeros:
    raiz = math.sqrt(n)
    if raiz.is_integer():
        enteros.append((n, int(raiz)))

print("Raices cuadradas exactas encontradas:")
for n, r in enteros:
    print(F"Raíz cuadrada de {n} = {r}")
````

10 - Haz un programa para una tienda que vende césped para jardines. Esta tienda trabaja con jardines circulares y el precio del metro cuadrado de césped es de R$ 25,00. Pide a la persona usuaria el radio del área circular y devuelve el valor en reales de cuánto tendrá que pagar.

````python
# Precio por metro cuadrado
precio_m2 = 25.00

# Solicita el radio del jardín
radio = float(input("Ingrese el radio del jardín (en metros): "))

# Calcula el área del círculo
area = math.pi * (radio ** 2)

# Calcula el costo total
costo_total = area * precio_m2

# Muestra el resultado
print(f"El area del jardin es de {area:.2f} m2")
print(f"El costo total del cesped será de R$ {costo_total:.2f}")
````
