8 - Crea un programa que solicite dos valores numéricos, un numerador y un denominador, y realice la división entre los dos valores. Asegúrate de que el valor del denominador no sea igual a 0.

```python
numerador = int(input("Ingrese un número (numerador)"));

while True:
    denominador = float(input("Ingresa un denominador distinto de cero: "))

    if denominador != 0:
        break
    print("Error: el denominador no puede ser 0. Intenta de nuevo.")

resultado = numerador/denominador
print(f"El resultado de la división de {numerador} entre {denominador} es {resultado}")
```

12 - Crea un código que solicite las 3 notas de un estudiante e imprima el promedio de las notas.

```python
notas = []

for i in range(3):
    nota = float(input(f"Ingrese nota #{i + 1}:"))
    print(f"Nota #{i + 1} = {nota}")
    notas.append(nota)

promedio = sum(notas) / len(notas)

print(f"El promedio de las notas del estudiante es {promedio:.1f}")
```

13 - Crea un código que calcule e imprima el promedio ponderado de los números 5, 12, 20 y 15 con pesos respectivamente iguales a 1, 2, 3 y 4.

```python
# Lista de números:
numeros = [5, 12, 20, 15]

# Lista con pesos correspondientes:
pesos = [1, 2, 3 , 4]


suma_ponderada = sum(num * peso for num, peso in zip(numeros, pesos))
suma_pesos = sum(pesos)

promedio_ponderado = suma_ponderada / suma_pesos

print(f"El promedio ponderado es: {promedio_ponderado}")
```
